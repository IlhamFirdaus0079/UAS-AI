```python
import pandas as pd

# Gantilah "nama_file.csv" dengan nama file dataset Anda
file_path = "Dataset6.csv"

# Membaca dataset
dataset = pd.read_csv(file_path)

# Menampilkan dataset
print("Dataset Awal Usaha Besar:")
print(dataset)

# Mencari pendapatan dari yang paling besar ke terkecil
dataset_sorted = dataset.sort_values(by='Usaha besar', ascending=False)

# Menampilkan dataset yang sudah diurutkan
print("\nDataset Usaha Besar Setelah Diolah Dan Diurutkan:")
print(dataset_sorted)
```

```python
import matplotlib.pyplot as plt

# Ganti 'Kategori' dan 'Jumlah' dengan kolom sesuai dataset Anda
labels = dataset['Usaha kecil']
sizes = dataset['Usaha kecil']

# Membuat diagram bundar
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Diagram Bundar dari Dataset Usaha kecil')

# Menampilkan diagram bundar
plt.show()
```

```python
import matplotlib.pyplot as plt

# Ganti 'Kategori' dan 'Jumlah' dengan kolom sesuai dataset Anda
labels = dataset['Usaha sedang']
sizes = dataset['Usaha sedang']

# Membuat diagram bundar
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Diagram Bundar dari Dataset Usaha sedang')

# Menampilkan diagram bundar
plt.show()
```
